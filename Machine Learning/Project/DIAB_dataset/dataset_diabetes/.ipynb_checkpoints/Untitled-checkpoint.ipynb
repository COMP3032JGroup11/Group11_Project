But there are still some null values in the data, and all the values in ICD-9 areless than 1000, so we let the null value be 1000. And we need to convert v|e into number, so we change them into 0.

```python
data['diag_1'].fillna('1000', inplace=True)
data['diag_2'].fillna('1000', inplace=True)
data['diag_3'].fillna('1000', inplace=True)
```

```python
def change_word(diag):
    if diag[0] =="V" or diag[0] =="E":
        return '0'
    else:
        return diag
```

```python
data['diag_1'] = data['diag_1'].apply(lambda diag:change_word(diag))
data['diag_2'] = data['diag_2'].apply(lambda diag:change_word(diag))
data['diag_3'] = data['diag_3'].apply(lambda diag:change_word(diag))
```

```python
def change_diag(diag):
    if float(diag)>=1 and float(diag)<140:
        return 0
    elif float(diag)>=140 and float(diag)<240:
        return 1
    elif float(diag)>=240 and float(diag)<280:
        return 2
    elif float(diag)>=280 and float(diag)<290:
        return 3
    elif float(diag)>=290 and float(diag)<320:
        return 4
    elif float(diag)>=320 and float(diag)<390:
        return 5
    elif float(diag)>=390 and float(diag)<460:
        return 6
    elif float(diag)>=460 and float(diag)<520:
        return 7
    elif float(diag)>=520 and float(diag)<580:
        return 8
    elif float(diag)>=580 and float(diag)<630:
        return 9
    elif float(diag)>=630 and float(diag)<680:
        return 10
    elif float(diag)>=680 and float(diag)<710:
        return 11
    elif float(diag)>=710 and float(diag)<740:
        return 12
    elif float(diag)>=740 and float(diag)<760:
        return 13
    elif float(diag)>=760 and float(diag)<780:
        return 14
    elif float(diag)>=780 and float(diag)<800:
        return 15
    elif float(diag)>=800 and float(diag)<1000:
        return 16
    elif float(diag)== 0:
        return 17
```

```python
data['diag_1'] = data['diag_1'].apply(lambda diag:change_diag(diag))
data['diag_2'] = data['diag_2'].apply(lambda diag:change_diag(diag))
data['diag_3'] = data['diag_3'].apply(lambda diag:change_diag(diag))
```

## Feature Selection

Then we will start feature selection.

### Univariate Selection

First, start with Univariate Selection

Univariate feature selection works by selecting the best features based on univariate statistical tests. Sci-kit learn has the SelectKBest function that uses the univariate selection to select the best K features out of the feature(>K) set you have.Let's see what are our best  features using chi-squared statistical test.

```python
target = data['readmitted']
data1 = data.drop('readmitted',axis=1)
best_features = SelectKBest(score_func=chi2, k=44).fit(data1, target)

datascores = pd.DataFrame(best_features.scores_)
datacolumns = pd.DataFrame(data1.columns)

#concat two dataframes for better visualization 
featureScores = pd.concat([datacolumns,datascores], axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns

print(featureScores.nlargest(10,'Score'))
```

### Tree-based feature selection

Then we use Tree-based feature selection. Tree-based estimators (see the sklearn.tree module and forest of trees in the sklearn.ensemble module) can be used to compute impurity-based feature importances, which in turn can be used to discard irrelevant features.

Feature importance gives you a score for each feature of your data, the higher the score more important or relevant is the feature towards your output variable.

We will use Extra Tree Classifier for extracting  the top 10 features for the dataset.

```python
tree = ExtraTreesClassifier()
tree.fit(data1, target)

df_feature_importances = pd.Series(tree.feature_importances_, index=data1.columns)
df_feature_importances.nlargest(10).plot(kind='barh')
```

Now we can train the KNN algorithm on this data and make predictions. Scikit-Learn has the sklearn.neighbors library, which contains built-in classes/methods for various k-nearest neighbors algorithms. You can find all the algorithm implementations here.

We have imported the KNeighborsClassifier class from the sklearn.neighbors library already. Let's initialize an instance of it with the most basic parameter which is the value for the K. There is no ideal value for K and it is selected after testing and evaluation, we select50.

Let's train this model using the normalized dataset and evaluate this model using test set.

```python
d_k_accuracy = []
for i in range(1,9):
    X_train, X_valid, y_train, y_valid = train_test_split(X_train_valid_n, y_train_valid, test_size=0.1*i,random_state=42)
    X_valid_n = sc.transform(X_valid)
    knn = KNeighborsClassifier(n_neighbors=50)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_valid_n)
    d_k_accuracy.append(metrics.accuracy_score(y_valid, pred))
```

## Single Classifier : Decision Tree

### Validation

Now we can train the decision tree algorithm on this data and make predictions. Scikit-Learn contains the tree library, which contains built-in classes/methods for various decision tree algorithms.

Since we are going to perform a classification task here, we will use the DecisionTreeClassifier class.

```python
d_dt_accuracy = []
for i in range(1,9):
    X_train, X_valid, y_train, y_valid = train_test_split(X_train_valid, y_train_valid, test_size=0.1*i,random_state=42)
    tree = DecisionTreeClassifier(criterion = 'entropy',random_state=42)
    tree.fit(X_train, y_train)
    pred = tree.predict(X_valid)
    d_dt_accuracy.append(metrics.accuracy_score(y_valid, pred))
```

```python
plt.figure(figsize=(20, 6))
plt.plot(range(1, 9), d_dt_accuracy, color='red', marker='o',
         markersize=10)
plt.title('Accuracy Vs Validation Hyperparameters')
plt.xlabel('Validation Hyperparameters')
plt.ylabel('Accuracy')
```

We select test_size = 0.1

### Hyperparameters Evaluation

#### maxdepth and  criterion

```python
tree_edt_accuracy = []
tree_gdt_accuracy = []
X_train, X_valid, y_train, y_valid = train_test_split(X_train_valid, y_train_valid, test_size=0.1,random_state=42)
for i in range(1,7):
    tree = DecisionTreeClassifier(criterion = 'entropy',max_depth=i,random_state=42)
    tree.fit(X_train, y_train)
    pred = tree.predict(X_valid)
    tree_edt_accuracy.append(metrics.accuracy_score(y_valid, pred))
    print("Entropy" , i, metrics.classification_report(y_valid, pred))
for i in range(1,7):
    tree = DecisionTreeClassifier(criterion = 'gini',max_depth=i,random_state=42)
    tree.fit(X_train, y_train)
    pred = tree.predict(X_valid)
    tree_gdt_accuracy.append(metrics.accuracy_score(y_valid, pred))
    print("Gini" , i  , metrics.classification_report(y_valid, pred))
```

```python
plt.figure(figsize=(20, 6))
plt.plot(range(1, 7), tree_edt_accuracy, color='red', marker='o',
         markersize=10)
plt.plot(range(1, 7), tree_gdt_accuracy, color='green', marker='o',
         markersize=10)
plt.title('Accuracy Vs depth & criterion')
plt.xlabel('depth')
plt.ylabel('Accuracy')
```
